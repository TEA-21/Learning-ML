```python
import matplotlib.pyplot as plt
import numpy as np


k=int(input('Enter a number: '))
n=list((np.arange(3,k)))
m=[]
seq= []
for i in n:
    while i > 1:
      seq= []
      seq.append(i)
      if i%2==0:
        i/=2
      else:
        i=3*1+1
    m.append(max(seq))
plt.figure(figsize=(13,7))
plt.scatter(n,m)
plt.show()
```
