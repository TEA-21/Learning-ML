```python
import os
import json

# 1. SET YOUR PATHS
# Use the path you copied from the sidebar here:
source_folder = ''
target_folder = '/content/drive/MyDrive/Notebooks_For_AI_Studio'

if not os.path.exists(target_folder):
    os.makedirs(target_folder)

print(f"Checking: {source_folder}")

# 2. CONVERT .ipynb TO .txt
files = [f for f in os.listdir(source_folder) if f.endswith('.ipynb')]
print(f"Found {len(files)} notebooks. Starting conversion...\n")

for filename in files:
    try:
        source_path = os.path.join(source_folder, filename)
        target_path = os.path.join(target_folder, filename.replace('.ipynb', '.txt'))

        # Open and read the notebook JSON
        with open(source_path, 'r', encoding='utf-8') as f:
            data = json.load(f)

        # Extract code and markdown only
        readable_text = ""
        for cell in data.get('cells', []):
            cell_type = cell.get('cell_type', 'unknown')
            source = "".join(cell.get('source', []))
            readable_text += f"--- {cell_type.upper()} CELL ---\n{source}\n\n"

        # Save as a clean .txt file
        with open(target_path, 'w', encoding='utf-8') as f:
            f.write(readable_text)

        print(f"✅ Prepared: {filename.replace('.ipynb', '.txt')}")

    except Exception as e:
        print(f"❌ Failed to convert {filename}: {e}")

print(f"\nDone! Now you can upload everything in '{target_folder}' to AI Studio.")
```
